```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('StockNews.csv')
df.head()
```

```python
df.shape
```

```python

# Drop the unnamed index column
df.drop(columns=["Unnamed: 0"], inplace=True)


# Rename the remaining column to 'Headline'
df.columns = ['Headline']

# Reset the index (optional)
df.reset_index(drop=True, inplace=True)

# Display the cleaned DataFrame
print(df.head())
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to calculate sentiment scores
def analyze_sentiment(News):
    return analyzer.polarity_scores(News)['compound']

# Apply sentiment analysis on headlines
df['Sentiment Score'] = df['Headline'].apply(analyze_sentiment)
# Classify sentiment based on the compound score
df['Sentiment'] = df['Sentiment Score'].apply(lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral'))
# Display the DataFrame with sentiment scores
df
```
